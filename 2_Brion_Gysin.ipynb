# Hierarchisierung der Zeichen bei Daten <font color="blue">(Texten):</font>
**EBENE 0** - auf niedrigster Ebene existieren nur **0 und 1**.<br>
**EBENE 1** - durch Kombinationen von 0 und 1 werden **Buchstaben, Ziffern, Sonderzeichen** etc. codiert.<br>
**EBENE 2** - durch Kombinationen von **Buchstaben** werden **Wörter** (mit Ziffern Zahlen) codiert.<br>
**EBENE 3** - durch Kombinationen von **Wörtern** werden **Sätze** codiert.<br>
**EBENE 4** - durch Kombinationen von **Sätzen** werden **Absätze** codiert.<br>

# Hierarchisierung der Zeichen bei Programmen <font color="blue">(Algorithmen):</font>
**EBENE 0, 1 und 2** sind identisch. <br>
**EBENE 3** - es gibt unterschiedliche Typen von Wörtern (Key-words, Kommentare, Variablennamen, Funktionen, etc.)<br>
**EBENE 4** - Modularisierung und Hierarchisierung von (Programmen, Bibliotehken, Funktionen, etc.)<br>

# Landsberger Poesieautomat, Hans Magnus Enzensberger, 1974

Beispiel der Kombination auf Wortebene.

In jeder **Zeile 6 Zufallselemente** aus einer Menge von **10 Zufallselementen**. Davon **6 Zeilen**, die ebenso aufgebaut sind. Das ergibt Zehn hoch 6 hoch 6 = 10 hoch 36 **Variationen** (zweiten Grades).

*Brühwarme Andacht <font color="blue">unter Zeitdruck</font>. Dieser gelehrige <font color="magenta">Edelmut</font> vor dem Sturm.<br>
Und diese sinnreichen <font color="cyan">Orgasmen</font>: Purer Zufall! Dabei gelingt uns das schon.<br>
Einstweilen schnell noch zähflüssige Schlussrunden. Grundsätzlich sparen!<br>
Sprechstunden. („Deine Freunde sind eben so neurotisch.“) An der Basis <font color="red">dumpfe Wut</font>.<br>
Die eiserne Stechuhr sagt mehr über uns als die Vernunft. Neuerdings dichten wir eben.<br>
Ausbrüche. Ratlosigkeit. Abblätternde <font color=#800080> Paradiese</font>. Im Zweifelsfall sind wir dran.*<br>

*Tropfenweise Hoffnung <font color="blue">unter Zeitdruck</font>. Dieser staubige <font color="magenta">Edelmut</font> vor der Pleite.<br>
Und diese vorgedruckten Zahlungsbefehle. Das kennt man! Freilich verlangen wir immer irgendwas.<br>
Danach auf Antrag schrille Proteste. Ungerührt sparen!<br>
Telefonspitzel. („Das Publikum war immer so gierig!“) Im Überbau <font color="red">dumpfe Wut</font>.<br>
Der graue Anzeigenteil ist zählebig wie Impotenz. Andrerseits schuften wir ganz allein.<br>
Sachzwänge, Biergeruch, wuchernde Weichbilder! <font color="green">In kleinen Mengen</font> fließt Blut.*<br>

*Vorbildliche Liebkosungen zum Frühstück. Dieser geizige Lustgewinn auf der Bühne.<br>
Und die langatmigen <font color="cyan">Orgasmen</font>. Beneidenswert! Übrigens ertragen wir das schon.<br>
Stattdessen wie immer weiche Zitate. Unter Tränen ausflippen!<br>
Beteuerungen. („Ich war aber so schlaff.“) Am Flughafen Kerzenlicht.<br>
Das schwärmende Andenken reut uns. Nicht die die (sic!) Grübelei. Anstandshalber schlagen wir erst mal.<br>
Spitzengespräche. Abfall. Fossile <font color=#800080>Paradiese</font>. <font color="green">In kleinen Mengen</font> nutzt alles nichts.*<br>

# Brion Gysin- Sound Poetry

[Sound Poetry - I AM THAT I AM](https://www.youtube.com/watch?v=RnMKF5Zyc-0)

Permutation aller Wörter des Satzes "I AM THAT I AM" (1960).

<br>
<br>


```python
# write python code that generates all permutations of the sentence 'I am what I am'.
# print all the permutations.

import itertools

# create a list of the words in the sentence
sentence = ['I', 'am', 'what', 'I', 'am']

# create all permutations of the sentence
permutations = itertools.permutations(sentence)

# print each permutation
for permutation in permutations:
    print(' '.join(permutation))
    
```

```python
# Install a text-to-speech conversion library for Python
#!pip install pyttsx3
```

```python
# Write python code that generates all permutations of the sentence 'no poet's don't own words'. 

import itertools

sentence = 'no poet\'s don\'t own words'

for perm in itertools.permutations(sentence.split()):
    print(' '.join(perm))

# write a speech synthesiser that speaks all the generated permutations.

import pyttsx3

engine = pyttsx3.init()

for perm in itertools.permutations(sentence.split()):
    engine.say(' '.join(perm))

engine.runAndWait()
```
